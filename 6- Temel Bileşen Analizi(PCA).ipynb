- Temek fikir, çok değişkenli verinin ana özelliklerini daha az sayıda değişken/bileşen ile temsil etmektir. Örneğin elimizde 100 tane değişken olduğunu düşünelim. Bu 100 tane değişkenin taşıdığı bilgiyi ondan daha az sayıda değişken (ör: 2 tane) ile temsil etmeye çalışıyoruz. Diğer bir ifade ile küçük miktarda bilgi kaybını göze alıp değişken boyutunu azaltıyoruz.




- Eğer elinizdeki 100 değişkenden ve gözlem birimlerinden oluşan veri setini kümeleme yöntemiyle kümeledikten sonra kümeleri görselleştirmek istiyorsanız, veri setinizi temel bileşenler analizi ile iki boyuta indirgeyip kümeleme yaklaşımını görelleştirebilirsiniz.
- Temel Bileşenler Analizi, genel olarak görüntü işleme ve regresyon modellerinde ortaya çıkan bazı problemlerde kullanılmaktadır. Özellikle regresyon modellerinde veri setinde çok fazla değişken olduğunda değişkenlerden bazıları birbiri ile ilişkili ise ortaya <span style="color: red">çoklu doğrusal bağlantı problemi</span> çıkar

- Teknik olarak arkada nasıl bir işlem gerçekleşiyor da çok boyutlu veri seti 2-3-4 ve benzeri bileşenlere indirgeniyor?
    - Co-varyans ya da korelasyon matrislerinin özdeğer ve özvektörlerinin bulunması aracılığıyla gerçekleştiriliyor. Değişken gruplarının varyanslarını ifade eden özdeğerler ile veri setindeki değişkenler gruplandırılıyor. Gruplar arasında en fazla varyansa sahip gruplar en önemli gruplar oluyor. Bu gruplara da biz <span style="color: red">temel(asal) bileşen</span> diyoruz.
