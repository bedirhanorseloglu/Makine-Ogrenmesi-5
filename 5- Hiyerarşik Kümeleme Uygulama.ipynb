```python
# Kütüphaneler

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram

import warnings
warnings.filterwarnings("ignore" , category=DeprecationWarning)
warnings.filterwarnings("ignore" , category=FutureWarning)
```

```python
# Veri setimiz (Amerika'daki bazı suç istatistiklerini ifade ediyor.)

df = pd.read_csv("USArrests.csv" , index_col=0)
df.head()

# Amacımız bu eyaletleri segmentlere ayırmak
```

# MODEL

```python
hc_complate = linkage(df , "complete")     # dendogram oluşturmak için kullanacağız
hc_average = linkage(df , "average")     # dendogram oluşturmak için kullanacağız
```

<h3 style="color: pink">Complete Yöntemi</h3>

```python
# Şimdi bunu görselleştirelim
plt.figure(figsize=(15,10))        # çıkacak olan görselin boyutunu ayarlamak için "figure" kullanıyoruz
plt.title("Hiyerarşik Kümeleme Dendogramı")
plt.xlabel("Gözlem Birimleri")
plt.ylabel("Hesaplanan Uzaklık")

dendrogram(hc_complate , leaf_font_size=10);     # "leaf_font_size" yazı boyutuyla alakalı bir parametre (gözlem birimleri ekseninindeki yazıların boyutu)
                                                # 7. satırın sonuna ";" eklersek çıktıdaki grafiğin üzerinde yazanlar gözükmez
```

```python
# Örneğin 10 kümeye ayırmaya karar verdik. Peki ben bunu 4 kümeye ayırdığımda bu gözlemleri nasıl gözürüz? Hangi kümede kaç tane gözlem var?


# Şimdi bunu görselleştirelim
plt.figure(figsize=(15,10))        # çıkacak olan görselin boyutunu ayarlamak için "figure" kullanıyoruz
plt.title("Hiyerarşik Kümeleme Dendogramı")
plt.xlabel("Gözlem Birimleri")
plt.ylabel("Hesaplanan Uzaklık")

dendrogram(hc_complate , leaf_font_size=10 , truncate_mode="lastp" , p=10 , show_contracted=True);     # "truncate_mode" -> en son(last) p (küme sayımız) adet göster
                                                                                                      # "show_contracted" -> kümeleme yapıldığında hangi kümede kaçr tane gözlem olduğunu göster
```

<h3 style="color: pink">Average Yöntemi</h3>

```python
plt.figure(figsize=(15,10))        
plt.title("Hiyerarşik Kümeleme Dendogramı")
plt.xlabel("Gözlem Birimleri")
plt.ylabel("Hesaplanan Uzaklık")

dendrogram(hc_average , leaf_font_size=10 , truncate_mode="lastp" , p=10 , show_contracted=True);     
```
