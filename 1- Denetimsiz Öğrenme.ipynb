# DENETİMSİZ ÖĞRENME

- Denetimsiz öğrenme, elimizdeki veri setinde bağımlı değişken olmadığında yani bağımsız değişkenler belirli bir çıktı ile ilişkilendirilmediğinde oluşan veri setlerinin üzerinde yapılan öğrenme işlemidir.



***Bu Konu Kapsamında Ele Alacağımız Konular***
- K-Means
- Hiyerarşik Kümeleme Analizi
- Temel Bileşen Analizi
